```python
import numpy as np
import math
import matplotlib.pyplot as plt
from decimal import Decimal
import time
```

```python
def get_div_modulo(p, val):
    if p > val:
        val = val * 10
    remainder = val%p
    val_debug = val//p
    return remainder
        

def digits_until_repetition(p):
    remainder = 1
    remainder_list = []
    we_are_done = False
    while not we_are_done:
        remainder = get_div_modulo(p, remainder)
        if remainder in remainder_list:
            we_are_done = True
        else:
            remainder_list.append(remainder)
    return len(remainder_list)
        
```

```python
#Taken from https://stackoverflow.com/questions/2068372/fastest-way-to-list-all-primes-below-n/3035188#3035188
def primesfrom3to(n):
    """ Returns a array of primes, 3 <= p < n """
    sieve = np.ones(n//2, dtype=bool)
    for i in range(3,int(n**0.5)+1,2):
        if sieve[i//2]:
            sieve[i*i//2::i] = False
    return 2*np.nonzero(sieve)[0][1::]+1
```

```python
our_prime_list = [2] + list(primesfrom3to(10000))
```

```python
# We print out computation times every 20 primes.
our_repeat_list = []
t0 = time.time()
for i, iprime in enumerate(our_prime_list):
    our_repeat_list.append(digits_until_repetition(iprime))
    if i%20 == 0:
        print(f"checking {iprime} at {time.time() - t0:4.2f} seconds")
```

```python
print(len(our_repeat_list))
```

```python
plt.figure(figsize = (15,10))
plt.scatter(our_prime_list, our_repeat_list)
plt.xlabel("Prime Number")
plt.ylabel("Digits to Repetition")
plt.show()
```
